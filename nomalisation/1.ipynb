```python
data = [1,4,5,6,7,8]
min_value = min(data);
max_value = max(data);
normalised = [(x-min_value)/(max_value-min_value)for x in data]
print(normalised)
```

```python
import numpy as np
data = np.array([10, 20, 30, 40, 50])
normalized = (data - data.min()) / (data.max() - data.min())
print(normalized)
```

```python
data = np.array(10, 20, 30, 40, 50)

mean = sum(data) / len(data)
std = (sum((x - mean) ** 2 for x in data) / len(data)) ** 0.5

normalized = [(x - mean) / std for x in data]

print(normalized)
```

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50])

normalized = (data - data.mean()) / data.std()

print(normalized)
```
